```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Importing Data
```

```python
df = pd.read_csv("Shark Tank India (impure).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
## Changing data type of date columns to datetime 

def data_type_to_date(date_columns):
    for i in date_columns:
        df[i] = pd.to_datetime(df[i])
```

```python
date_columns = ['Original_Air_Date','Season_End','Season_Start']
data_type_to_date(date_columns)
```

```python
df[date_columns].info()
```

```python
## Dropping Irrelevant column
df.drop(columns=['Company_Website','Original_Air_Date','Episode_Title','EBITDA','Cash_Burn','SKUs','Gross_Margin','Net_Margin'],inplace = True)
```

```python
## Handling Missing values
```

```python
df.isnull().sum()
```

```python
df['Started_in'] = df['Started_in'].fillna("Unknown")
df['Male_Presenters'] = df['Male_Presenters'].fillna(0)
df['Female_Presenters'] = df['Female_Presenters'].fillna(0)
df['Transgender_Presenters'] = df['Transgender_Presenters'].fillna(0)
df['Couple_Presenters'] = df['Couple_Presenters'].fillna(0)
df['Pitchers_City'] = df['Pitchers_City'].fillna("Unknown")
df['Pitchers_State'] = df['Pitchers_State'].fillna("Unknown")
df['Accepted_Offer']= df['Accepted_Offer'].fillna("No Offer Received")
df['Has_Patents'] = df['Has_Patents'].fillna(0)
df['Bootstrapped'] = df['Bootstrapped'].fillna(0)
df['Total_Deal_Amount(in_lakhs)']= df['Total_Deal_Amount(in_lakhs)'].fillna(0)
df['Total_Deal_Equity(%)']= df['Total_Deal_Equity(%)'].fillna(0)
df['Number_of_Sharks_in_Deal']= df['Number_of_Sharks_in_Deal'].fillna(0)
df['Namita_Investment_Amount(in lakhs)']= df['Namita_Investment_Amount(in lakhs)'].fillna(0)
df['Vineeta_Investment_Amount(in_lakhs)']= df['Vineeta_Investment_Amount(in_lakhs)'].fillna(0)
df['Anupam_Investment_Amount(in_lakhs)']= df['Anupam_Investment_Amount(in_lakhs)'].fillna(0)
df['Aman_Investment_Amount(in_lakhs)']= df['Aman_Investment_Amount(in_lakhs)'].fillna(0)
df['Peyush_Investment_Amount((in_lakhs)']= df['Peyush_Investment_Amount((in_lakhs)'].fillna(0)
df['Amit_Investment_Amount(in_lakhs)']= df['Amit_Investment_Amount(in_lakhs)'].fillna(0)
df['Ashneer_Investment_Amount']= df['Ashneer_Investment_Amount'].fillna(0)
df['Namita_Present']=df['Namita_Present'].fillna("No")
df['Vineeta_Present']= df['Vineeta_Present'].fillna('No')
df['Anupam_Present']= df['Anupam_Present'].fillna('No')
df['Aman_Present']= df['Aman_Present'].fillna('No')
df['Peyush_Present']= df['Peyush_Present'].fillna('No')
df['Amit_Present']= df['Amit_Present'].fillna('No')
df['Ashneer_Present']= df['Ashneer_Present'].fillna('No')
```

```python
df['Yearly_Revenue(in_lakhs)'] = df['Yearly_Revenue(in_lakhs)'].fillna("Not Mentioned")
df['Monthly_Sales(in_lakhs)'] = df['Monthly_Sales(in_lakhs)'].fillna("Not Mentioned")
```

```python
df.isnull().sum()
```

```python
## Replacing character 0 with ' ' in columns - ['Anchor','Industry','Business_Description','Pitchers_State']
```

```python
def replace_character(cols):
    for i in cols:
        df[i] = df[i].str.replace('0',' ')
    return df[cols].info()
```

```python
cols = ['Anchor','Industry','Business_Description','Pitchers_State']
replace_character(cols)
```
